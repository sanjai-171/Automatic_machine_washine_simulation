```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
%matplotlib inline
```

```python
N=13#length of FIR filter
Omc=0.6*np.pi/2

#compute impulse response
k= np.arange(N)
hd=(Omc/np.pi)*np.sinc((k-(N-1)/2)*Omc/np.pi)

#Rectangular window
w1=np.ones(N)
w2=np.blackman(N)
#desired FIR impulse response
h1=hd*w1
h2=hd*w2
#Frequency response
Om, H1=sig.freqz(h1)
Om, H2=sig.freqz(h2)
```

```python
plt.figure(figsize=(10,5))
plt.stem(h1,use_line_collection=True)
plt.title('Impulse response')
plt.xlabel(r'$n$')
plt.ylabel(r'$h[n]$')
```

```python
plt.figure(figsize=(10,5)) 
plt.stem (h2, use_line_collection=True)
plt.title('Impluse response(blackman)')
plt.xlabel (r'$n$')
plt.ylabel(r'$h[n]$')
```

```python
plt. figure(figsize=(10,5))
plt.plot([0, Omc, Omc], [0, 0, -100], 'r-', label='desired') 
plt.plot(Om,20*np.log10(abs(H1)),label='Rect method')
plt.plot(Om,20*np.log10(abs(H2)),label='Blackman method')
plt.title('Magnitude response')
plt.xlabel(r'$Omega$')
plt.ylabel(r'$|H(e^{j\omega} | $ in dB')
plt.axis ([0, np.pi, -120,3])
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(10,5))
plt.plot([0,Om[-1]],[0,0],'r-',label='desired')
plt.plot(Om,np.unwrap (np.angle(H1)),label='Rect method') 
plt.plot (Om,np.unwrap(np.angle(H2)),label='Blackman method')
plt.title('Phase response')
plt.xlabel(r'$Omega$')
plt.ylabel(r'$\varphi(\omega$ in rad')
plt.legend() 
plt.grid()
```
